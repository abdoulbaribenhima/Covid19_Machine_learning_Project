```python
url2="https://coronavirus.politologue.com/coronavirus-italie.IT"
url3="https://coronavirus.politologue.com/coronavirus-chine.CN"
i= pd.read_html(url2)
it=i[0]
c= pd.read_html(url3)
ch=c[0]
it.to_excel('Italy_covid19.xlsx')
it.to_csv('Italy_covid19.csv')
ch.to_excel('China_covid19.xlsx')
ch.to_csv('China_covid19.csv')
```
