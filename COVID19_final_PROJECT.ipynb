```python
df_word=pd.read_csv(r"C:\Users\Abdouuulbari\Desktop\ELITECH\Python\PROJET COVID19\COVID-19-Time-Series-Forecasting-with-Data-Analysis-master\covid_19_data.csv"
                    ,parse_dates=["Last Update"])
df_word
```

```python
df_word.rename(columns={"ObservationDate":"Date","Province/State":"State","Country/Region":"Country"},inplace=True)
df_word.head()
```

```python
df_word.query('Country=="France"')
```

```python
confirmed=df_word.groupby("Date").sum()["Confirmed"].reset_index()
deaths=df_word.groupby("Date").sum()["Deaths"].reset_index()
recovered=df_word.groupby("Date").sum()["Recovered"].reset_index()
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=confirmed["Date"],y=confirmed["Confirmed"],mode='lines+markers',name="Confirmed",line=dict(color='blue',width=2)))
fig.add_trace(go.Scatter(x=deaths["Date"],y=deaths["Deaths"],mode='lines+markers',name="Deaths",line=dict(color='red',width=2)))
fig.add_trace(go.Scatter(x=recovered["Date"],y=recovered["Recovered"],mode='lines+markers',name="Recovered",line=dict(color='green',width=2)))
```

```python
from plotly.subplots import make_subplots
```

```python
fig=make_subplots(
      rows=2,cols=2,
    specs= [[{"secondary_y":True},{"secondary_y":True}],[{"secondary_y":True},{"secondary_y":True}]],
    subplot_titles=("France","Maroc","Italy","China"))
##### pour organiser l'affichage des graphs

fig.add_trace(go.Bar(x=dbd_global_France["Date"],y=dbd_global_France["T_Cas"],
                    marker=dict(color=dbd_global_France["T_Cas"],coloraxis="coloraxis")),1,1)

fig.add_trace(go.Bar(x=dbd_global_Maroc["Date"],y=dbd_global_Maroc["T_Cas"],
                    marker=dict(color=dbd_global_Maroc["T_Cas"],coloraxis="coloraxis")),1,2)

fig.add_trace(go.Bar(x=dbd_global_Italy["Date"],y=dbd_global_Italy["T_Cas"],
                    marker=dict(color=dbd_global_Italy["T_Cas"],coloraxis="coloraxis")),2,1)

fig.add_trace(go.Bar(x=dbd_global_China["Date"],y=dbd_global_China["T_Cas"],
                    marker=dict(color=dbd_global_China["T_Cas"],coloraxis="coloraxis")),2,2)





fig.update_layout(coloraxis=dict(colorscale='Bluered_r'),showlegend=False,title_text="Total Cases in 4 Countries")
fig.update_layout(plot_bgcolor='rgb(230,230,230)')
```
